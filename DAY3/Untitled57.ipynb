```python
import pandas as pd

df = pd.read_csv('used_cars_data.csv')
df.shape
```

```python
df.info()
```

```python
new_df = df.sample(n=50)  ## random sampling 
new_df.shape
```

```python
new_df.columns
```

```python
# drop some unwanted columns
# rename columns Kilometeres_Driven -> km
# drop duplicates
# Test with null values -> replace null values
```

```python
new_df.drop(columns=['S.No.','New_Price','Transmission','Owner_Type','Seats'],inplace=True)
```

```python
new_df.columns
```

```python
new_df.shape
```

```python
new_df.head(3)
```

```python
new_df.isnull().sum()
```

```python
import numpy as np
new_df['Price'].fillna(np.mean(new_df['Price']),inplace=True)
```

```python
new_df.isnull().sum()
```

```python
new_df.duplicated().sum()
```

```python
new_df.drop_duplicates(inplace=True)
```

```python
new_df.isnull().sum()
```

```python
new_df.columns
```

```python
new_df.rename(columns={'Kilometers_Driven':'KM','Fuel_Type':'Fuel'},inplace=True)
```
